```python
import pandas as pd
import os

# Function to process the text file and convert it into a pandas DataFrame
def process_transcript(file_path, audio_directory):
    # Read the file into a DataFrame
    df = pd.read_csv(file_path, sep="\t", header=None, names=["sentence_id", "sentence"])

    # Fill NaN values in 'sentence_id' with the previous non-NaN value
    df['sentence_id'] = df['sentence_id'].fillna(method='ffill')

    # Convert 'sentence_id' to integer to remove decimal points
    df['sentence_id'] = df['sentence_id'].astype(int)

    # Group by 'sentence_id' and join the sentences
    df_grouped = df.groupby('sentence_id')['sentence'].apply(' '.join).reset_index()

    # Add a column for the audio file path
    df_grouped['audio_file'] = df_grouped['sentence_id'].apply(lambda x: os.path.join(audio_directory, f"{x}.WAV"))

    return df_grouped


```

```python
# Base directory for the transcript files
transcript_base_directory = '/content/SCRIPT/'

# Base directory where the audio folders are located
audio_base_directory = '/content/WAVE/'

# List to hold dataframes from each speaker's transcript for both sessions
dfs = []

# Loop through the speaker numbers
for i in range(1, 11):
    speaker_id = f"SPEAKER{str(i).zfill(4)}"

    for session_num in range(2):  # 0 for SESSION0, 1 for SESSION1
        transcript_file_name = f"20{str(i).zfill(3)}{session_num}.TXT"  # Adjusted to new naming convention
        transcript_file_path = os.path.join(transcript_base_directory, transcript_file_name)
        audio_directory_path = os.path.join(audio_base_directory, speaker_id, f"SESSION{session_num}")

        print(f"Checking: {transcript_file_path}")  # Debug print

        if os.path.isfile(transcript_file_path):
            print(f"Processing file: {transcript_file_path}")  # Debug print

            df = process_transcript(transcript_file_path, audio_directory_path)
            dfs.append(df)

        else:
            print(f"File not found: {transcript_file_path}")  # Debug print

if not dfs:
    print("No dataframes were created. No files were processed.")
else:
    # Concatenate all the dataframes
    combined_df = pd.concat(dfs, ignore_index=True)
    print("Dataframes successfully concatenated.")
# Concatenate all the dataframes
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
from transformers import WhisperTokenizer
from transformers import WhisperFeatureExtractor

feature_extractor = WhisperFeatureExtractor.from_pretrained("openai/whisper-small")

tokenizer = WhisperTokenizer.from_pretrained("openai/whisper-small", language="English", task="transcribe")
```

```python
input_str = combined_df['sentence'][0]
labels = tokenizer(input_str).input_ids
decoded_with_special = tokenizer.decode(labels, skip_special_tokens=False)
decoded_str = tokenizer.decode(labels, skip_special_tokens=True)

print(f"Input:                 {input_str}")
print(f"Decoded w/ special:    {decoded_with_special}")
print(f"Decoded w/out special: {decoded_str}")
print(f"Are equal:             {input_str == decoded_str}")


```

```python
import pandas as pd
import wave
import numpy as np

# Function to get amplitude array from an audio file
def get_amplitude_array(file_path):
    with wave.open(file_path, 'rb') as wave_file:
        assert wave_file.getframerate() == 16000, "Unexpected sampling rate in the audio file"
        raw_audio = wave_file.readframes(wave_file.getnframes())
        amplitude_array = np.frombuffer(raw_audio, dtype=np.int16)

        # Normalize and convert to float32
        amplitude_array = amplitude_array.astype(np.float32) / 32768.0
    return amplitude_array

# Assuming 'combined_df' is your DataFrame and it has a column named 'audio_file_path'
# Apply the function to each row and store the result in a new column

combined_df['amplitude_array'] = combined_df['audio_file'].apply(get_amplitude_array)


# Now 'combined_df' contains a new column 'amplitude_array' with the amplitude data for each audio file
```

```python
from datasets import Dataset
combined_df = Dataset.from_pandas(combined_df)
combined_df = combined_df.train_test_split(test_size = 0.2)
combined_df
```

```python
def prepare_dataset(batch):
    audio = batch["amplitude_array"]

    # compute log-Mel input features from input audio array
    batch["input_features"] = feature_extractor(audio, sampling_rate=16000).input_features[0]

    # encode target text to label ids
    batch["labels"] = tokenizer(batch["sentence"]).input_ids


prepared_df = combined_df.map(prepare_dataset,
                              remove_columns = combined_df.column_names['train']
                            )
```
